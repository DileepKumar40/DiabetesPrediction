```python
colors = {0:'blue', 1:'red'}
```

```python
plt.scatter(G, AGE, c=Outcome.apply(lambda Outcome: colors[Outcome]))
plt.title('Glucose, Age vs outcome')
plt.show()
```

```python
plt.scatter(G, I, c=Outcome.apply(lambda Outcome:colors[Outcome]))
plt.title('Glucose, Insuline vs outcome')
plt.show()
```
